```python
import pandas as pd
e_SEB = pd.read_pickle('Environmentresults/SEB.pkl')
s_SEB = pd.read_pickle('Socialresults/SEB.pkl')
g_SEB = pd.read_pickle('Governanceresults/SEB.pkl')
```

```python
g_SEB
```

```python
e_SEB = e_SEB.rename(columns={"Similarity score with query":"Similarity score with query_E","Document name":"Year"})
#e_SEB = e_SEB.drop(columns=["Document no","Most similar words"])
s_SEB = s_SEB.rename(columns={"Similarity score with query":"Similarity score with query_S","Document name":"Year"})
#s_SEB = s_SEB.drop(columns=["Document no","Most similar words"])
g_SEB = g_SEB.rename(columns={"Similarity score with query":"Similarity score with query_G","Document name":"Year"})
#g_SEB = g_SEB.drop(columns=["Document no","Most similar words"])
s_SEB
```

```python
dfs = [e_SEB, s_SEB, g_SEB]
dfs = [df.set_index('Year') for df in dfs]
s = dfs[0].join(dfs[1:])
#s = s.drop(columns=['Most similar words in environment topic', 'Most similar words in social topic','Most similar words in governance topic','E_Index of document','S_Index of document','G_Index of document'])
s.index.names = ['Quarter document']
s.mean()
```

```python

s = s.assign(Year_number = s.index.str[21:25])
s = s.sort_values(by='Year_number')
s = s[['Year_number', 'Similarity score with query_E', 'Similarity score with query_S','Similarity score with query_G']]
s
```

```python
s.plot(x ='Year_number', y='Similarity score with query_E', kind = 'line',title = "SEB")
```

```python
s.plot(x ='Year_number', y='Similarity score with query_S', kind = 'line',title = "SEB")
```

```python
s.plot(x ='Year_number', y='Similarity score with query_G', kind = 'line',title = "SEB")
```
