```python
import pandas as pd
e_LHV = pd.read_pickle('Environmentresults/LHV.pkl')
s_LHV = pd.read_pickle('Socialresults/LHV.pkl')
g_LHV = pd.read_pickle('Governanceresults/LHV.pkl')
e_LHV
```

```python
e_LHV = e_LHV.rename(columns={"Similarity score with query":"Similarity score with query_E","Document name":"Year"})
#e_LHV = e_LHV.drop(columns="Document no")
s_LHV = s_LHV.rename(columns={"Similarity score with query":"Similarity score with query_S","Document name":"Year"})
#s_LHV = s_LHV.drop(columns="Document no")
g_LHV = g_LHV.rename(columns={"Similarity score with query":"Similarity score with query_G","Document name":"Year"})
#g_LHV = g_LHV.drop(columns="Document no")
s_LHV
```

```python
dfs = [e_LHV, s_LHV, g_LHV]
dfs = [df.set_index('Year') for df in dfs]
s = dfs[0].join(dfs[1:])
#s = s.drop(columns=['Most similar words in environment topic', 'Most similar words in social topic','G_Most similar words','E_Index of document','S_Index of document','G_Index of document'])
#s = s.rename(columns = {"G_Similarity score":"Similarity score in governance topic"}) 
s.mean()
```

```python

s = s.assign(Year_number =s.index.str[:4])
s = s.sort_values(by='Year_number')
s
```

```python
s.plot(x ='Year_number', y='Similarity score with query_E', kind = 'line',title = "LHV")
```

```python
s.plot(x ='Year_number', y='Similarity score with query_S', kind = 'line',title = "LHV")
```

```python
s.plot(x ='Year_number', y='Similarity score with query_G', kind = 'line',title = "LHV")
```
