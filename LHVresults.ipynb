```python
import pandas as pd
e_LHV = pd.read_pickle('Environmentresults/LHV.pkl')
s_LHV = pd.read_pickle('Socialresults/LHV.pkl')
g_LHV = pd.read_pickle('Governanceresults/LHV.pkl')
```

```python
e_LHV = e_LHV.rename(columns={"Similarity score": "E_Similarity score","Index of document": "E_Index of document","Most similar words": "E_Most similar words","Document name":"Year"})
s_LHV = s_LHV.rename(columns={"Similarity score": "S_Similarity score","Index of document": "S_Index of document","Most similar words": "S_Most similar words","Document name":"Year"})
g_LHV = g_LHV.rename(columns={"Similarity score": "G_Similarity score","Index of document": "G_Index of document","Most similar words": "G_Most similar words","Document name":"Year"})
g_LHV
```

```python
dfs = [e_LHV, s_LHV, g_LHV]
dfs = [df.set_index('Year') for df in dfs]
s = dfs[0].join(dfs[1:])
s = s.drop(columns=['Most similar words in environment topic', 'Most similar words in social topic','G_Most similar words','E_Index of document','S_Index of document','G_Index of document'])
s = s.rename(columns = {"G_Similarity score":"Similarity score in governance topic"}) 
s
```

```python

s = s.assign(Year_number =s.index.str[:4])
s = s.sort_values(by='Year_number')
s
```

```python
s.plot(x ='Year_number', y='Similarity score in environment topic', kind = 'line')
```

```python
s.plot(x ='Year_number', y='Similarity score in social topic', kind = 'line')
```
