```python
import pandas as pd
e_swed = pd.read_pickle('Environmentresults/Swedbank.pkl')
s_swed = pd.read_pickle('Socialresults/Swedbank.pkl')
g_swed = pd.read_pickle('Governanceresults/Swedbank.pkl')
g_swed
```

```python
e_swed = e_swed.rename(columns={"Similarity score with query":"Similarity score with query_E","Document name":"Year"})
#e_swed = e_swed.drop(columns=["Document no","Most similar words"])
s_swed = s_swed.rename(columns={"Similarity score with query":"Similarity score with query_S","Document name":"Year"})
#s_swed = s_swed.drop(columns=["Document no","Most similar words"])
g_swed = g_swed.rename(columns={"Similarity score with query":"Similarity score with query_G","Document name":"Year"})
#g_swed = g_swed.drop(columns=["Document no","Most similar words"])
s_swed
```

```python
dfs = [e_swed, s_swed, g_swed]
dfs = [df.set_index('Year') for df in dfs]
s = dfs[0].join(dfs[1:])
s.index.names = ['Quarter document']
s.mean()
```

```python

s = s.assign(Year_number = "20" + s.index.str[3:5])
s = s.sort_values(by='Year_number')
s
```

```python
s.plot(x ='Year_number', y='Similarity score with query_E', kind = 'line', title = "SWEDBANK")
```

```python
s.plot(x ='Year_number', y='Similarity score with query_S', kind = 'line', title = "SWEDBANK")
```

```python
s.plot(x ='Year_number', y='Similarity score with query_G', kind = 'line', title = "SWEDBANK")
```
